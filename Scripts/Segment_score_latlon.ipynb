```python
path = "/Users/poker/Desktop/DVA_Group_Project/Metric/Safe_score_segment.csv"
data = pd.read_csv(path)
data.head()
```

```python
score = data[["cnn_sgmt_pkey", "f_node_cnn_intrsctn_fkey", "t_node_cnn_intrsctn_fkey", "Scale_Score_y2"]]
```

```python
score.head()
```

```python
path_1 = "/Users/poker/Desktop/DVA_Group_Project/Database/TransBase/geo_st_intrsctn.csv"
latlon = pd.read_csv(path_1)
```

```python
latlon = latlon[["cnn_intrsctn_pkey", "latitude", "longitude"]]
latlon.head()
```

```python
new = score.merge(latlon, left_on = "f_node_cnn_intrsctn_fkey", right_on = "cnn_intrsctn_pkey", how = "inner")
new.head()
```

```python
new.drop(columns = ["cnn_intrsctn_pkey"], inplace = True)
```

```python
new_all = new.merge(latlon, left_on = "t_node_cnn_intrsctn_fkey", right_on = "cnn_intrsctn_pkey", how = "inner")
new_all.head()
```

```python
segment_latlon = new_all.drop(columns = ["cnn_intrsctn_pkey"])
segment_latlon
```

```python
segment_latlon = segment_latlon.drop(columns = ["f_node_cnn_intrsctn_fkey", "t_node_cnn_intrsctn_fkey"])
segment_latlon
```

```python
import csv
import json
from collections import OrderedDict
```

```python
li = []

for index, row in segment_latlon.iterrows():
    d = OrderedDict()
    d['type'] = 'Feature'
    d['geometry'] = {
            'type': "LineString",
            'coordinates': [
                [float(row["longitude_x"]), float(row["latitude_x"])],
                [float(row["longitude_y"]), float(row["latitude_y"])]
            ]
        }
    d['properties'] = {
            'score': row["Scale_Score_y2"],
            'sgmt': row['cnn_sgmt_pkey']
        }
    li.append(d)
```

```python
d = OrderedDict()
d['type'] = 'FeatureCollection'
d['features'] = li
```

```python
d
```

```python
with open('GeoObs.json', 'w') as f:
    f.write(json.dumps(d, sort_keys=False, indent=4))
```
